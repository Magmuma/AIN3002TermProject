```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
from tensorflow.keras.layers import Dropout
from tensorflow.keras.utils import to_categorical

```

```python
(train_images, train_labels), (test_images, test_labels) = datasets.cifar100.load_data()

# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0

# Convert class vectors to binary class matrices.
train_labels = to_categorical(train_labels, 100)
test_labels = to_categorical(test_labels, 100)
```

```python
model1 = models.Sequential()
model1.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model1.add(layers.MaxPooling2D((2, 2)))
model1.add(layers.Conv2D(128, (3, 3), activation='relu'))
model1.add(layers.MaxPooling2D((2, 2)))
model1.add(layers.Conv2D(256, (3, 3), activation='relu'))
model1.add(layers.MaxPooling2D((2, 2)))
model1.add(layers.Flatten())
model1.add(layers.Dense(256, activation='relu'))
model1.add(layers.Dense(100, activation='softmax'))

model1.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=['accuracy'])

history1 = model1.fit(train_images, train_labels, epochs=50, 
                    validation_data=(test_images, test_labels))


```

```python
# Model 2: ConvNet + Max Pooling + Dropout
model2 = models.Sequential()
model2.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model2.add(layers.MaxPooling2D((2, 2)))
model2.add(layers.Conv2D(128, (3, 3), activation='relu'))
model2.add(layers.MaxPooling2D((2, 2)))
model2.add(layers.Conv2D(256, (3, 3), activation='relu'))
model2.add(layers.MaxPooling2D((2, 2)))
model2.add(layers.Flatten())
model2.add(layers.Dense(256, activation='relu'))
model2.add(layers.Dropout(0.5))  # Dropout layer
model2.add(layers.Dense(100, activation='softmax'))

model2.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=['accuracy'])

history2 = model2.fit(train_images, train_labels, epochs=50, 
                    validation_data=(test_images, test_labels))


```

```python
# Calculate the error rate
error_rate1 = 1 - history1.history['val_accuracy'][-1]
print('Error rate for model1:', error_rate1*100)

# Calculate the error rate with dropout
error_rate2 = 1 - history2.history['val_accuracy'][-1]
print('Error rate for model2:', error_rate2*100)
```
